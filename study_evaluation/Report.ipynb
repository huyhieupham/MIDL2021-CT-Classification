```python
import  pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import scipy.stats
```

```python
def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m+h, m-h
```

```python
bootstrap = pd.read_csv('/home/single1/BACKUP/binhdao/binhdao/study_evaluation/bootstrap_f1.csv')
key1 = 'Percentage of slices for study level prediction'
data = {'Accuracy': [], key1: []}
keys = list(bootstrap.columns.values)

for i, row in bootstrap.iterrows():
    for key in keys:
        data[key1].append(int(key))
        data['Accuracy'].append(row[key])  
df = pd.DataFrame(data)
df.columns = ['F1-score', 'Percentage of slices for study level prediction']

ax = sns.lineplot(x=df[key1], y=df['F1-score'], ci=95)
plt.savefig('95_CI.png')
print(mean_confidence_interval(df[df['Percentage of slices for study level prediction'] == 30]['F1-score']))
```
